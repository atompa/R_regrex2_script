```r
csv_data<-read.csv("regrex2.csv")
print(csv_data)
```

```r
plot(x = csv_data$x,
     y = csv_data$y,
     xlab = "x",
     ylab = "y",
     main = "Regrex Data")
```

```r
plot(x = csv_data$x,
     y = csv_data$y,
     xlab = "x",
     ylab = "y",
     col = "black",
     main = "Regrex Data")
  
abline(lm(csv_data$y ~ csv_data$x,
          data = csv_data), col = "black")
```
